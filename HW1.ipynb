## Задание 2
Написать скрипт, который меняет местами луну и голубую воронку (в самом центре) на изображении starry_night.jpg.

```python
img = cv.imread('starry_night.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

funnel = img[110:325, 285:480, :]
moon = img[35:185, 600:730, :]

funnel = cv.resize(funnel, funnel.shape[:2])
moon = cv.resize(moon, moon.shape[:2])

img[145:(145 + moon.shape[0]), 310:(310 + moon.shape[1]), :] = moon
img[20:(20 + funnel.shape[0]), 525:(525 + funnel.shape[1]), :] = funnel

plt.imshow(img)
```

## Задание 3
Попробуйте превратить "starry_night" в "starry_day" путем увеличения яркости изображения и смены темно-синих тонов на светло-желтые.

```python
img = cv.imread('starry_night.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
def change_rgb(img, red=1.7, green=0.8, blue=1.5):

  img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

  img_transformed = img_rgb * [blue, green, red]
  img_transformed = np.clip(img_transformed, 0, 255).astype(np.uint8)

  img = cv.cvtColor(img_transformed, cv.COLOR_BGR2RGB)
  return img
```

```python
img_res = change_rgb(img, 2.7, 1.8, 0.7)
plt.imshow(img_res)
```
