```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
from math import sin, cos, pi
```

## Задание 1
Повернуть изображение относительно его центра (а не начала координат).

```python
img = cv.imread('messi5.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
height, width = img.shape[:2]
cx, cy = width // 2, height // 2

angle = 30 * pi/ 180
M = np.float32([
    [cos(angle), -sin(angle), (1 - cos(angle)) * cx + sin(angle) * cy],
    [sin(angle), cos(angle),  (1 - cos(angle)) * cy - sin(angle) * cx]]
)

img_res = cv.warpAffine(img, M, (width, height))

plt.imshow(img_res)
```

## Задание 2
Используя аффинные преобразования, необходимо "выправить" изображения (привести в исходный вид).

```python
img = cv.imread('apple.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
height, width = img.shape[:2]
cx, cy = width // 2, height // 2

angle = pi
M = np.float32([
    [cos(angle), -sin(angle), (1 - cos(angle)) * cx + sin(angle) * cy],
    [sin(angle), cos(angle),  (1 - cos(angle)) * cy - sin(angle) * cx]]
)

img_res = cv.warpAffine(img, M, (width, height))

plt.imshow(img_res)
```

```python
img = cv.imread('skull.png')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
height, width = img.shape[:2]

src_points = np.float32([[0, 0], [width-1, 0], [0, height-1]])
dst_points = np.float32([[0, 0], [int(0.6*(width-1)), 0], [int(0.4*(width-1)), height-1]])

M = cv.getAffineTransform(src_points, dst_points)

img_affine = cv.warpAffine(img, M, (width, height))

img_res = cv.resize(img_affine, dsize=(int(1.5 * width), height), interpolation=cv.INTER_CUBIC)
cropped_img_res = img_res[165:900, 1100:1600]

plt.imshow(cropped_img_res)
```
