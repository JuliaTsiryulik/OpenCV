```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
from math import sin, cos, pi
```

## Задание 1
Используя модель HSV выполнить задание по превращению "starry_night" в "starry_day".

```python
img = cv.imread('starry_night.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
def change_hsv(img, value=50, hue=60):
  hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
  h, s, v = cv.split(hsv)

  h = cv.add(h,hue)
  v = cv.add(v,value)

  v[v > 255] = 255
  v[v < 0] = 0
  h[h > 255] = 255
  h[h < 0] = 0

  final_hsv = cv.merge((h, s, v))
  img = cv.cvtColor(final_hsv, cv.COLOR_HSV2BGR)
  return img
```

```python
img_res = change_hsv(img, 60, 80)

plt.imshow(img_res)
```

## Задание 2
Имея размытое изображение, как сделать его более четким? Попробуйте подобрать матрицу, которая может это сделать.

```python
img = cv.imread('Penguins.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

img_blur = cv.GaussianBlur(img, (9,9), 0)
plt.imshow(img_blur)
```

```python
sharpen_kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
img_res = cv.filter2D(img_blur, -1, sharpen_kernel)
plt.imshow(img_res)
```
